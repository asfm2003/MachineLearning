Note: We won't use the loss function directly for this scenerio as we have to calculate derivatives for it. So we can use it independently to get the value of Error. For Gradietn Descent we will is this formula separately rather than calling this function.

# Gradient Descent:

```python
def gradient_desc(m_now,b_now,points,L):
    m_gradient = 0
    b_gradient = 0
    n = len(points)
    
    for i in range(n):
        x = points.iloc[i].study_hours
        y = points.iloc[i].scores
        
        m_gradient += (-2/n) * x *(y-(m_now*x + b_now))
        b_gradient += (-2/n) * (y-(m_now*x + b_now))
    
    m = m_now - m_gradient*L
    b = b_now - b_gradient*L
    return m,b
```

# Testing Linear Regression

```python
m = 0
b = 0
L = 0.0001
epochs = 250 
#epochs define the number of times an algo would iterate

for i in range(epochs):
    m,b = gradient_desc(m,b,data,L)
print(m,b)
```
